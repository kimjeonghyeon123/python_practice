```python
ratings = pd.read_csv('/content/gdrive/MyDrive/ratings_small.csv')
movies = pd.read_csv('/content/gdrive/MyDrive/movies_metadata.csv')

ratings['movieId'] = ratings['movieId'].astype(str)
movies['id'] = movies['id'].astype(str)

movies.rename(columns = {'id': 'movieId'}, inplace = True)

ratings_movies = pd.merge(ratings, movies, on = 'movieId')

ratings_movies.head(1)
```

```python
ratings_movies.shape
```

```python
data = ratings_movies.pivot_table('rating', index = 'userId', columns = 'original_title').fillna(0)
```

```python
data.head()
```

```python
data.shape
```

```python
data = data.transpose()
data.head(2)
```

```python
data.shape
```

```python
movie_sim = cosine_similarity(data, data)
print(movie_sim.shape)
```

```python
movie_sim_df = pd.DataFrame(data = movie_sim, index = data.index, columns = data.index)
```
